#Este es un simulador que muestra una caja registradora de una tienda de barrio. Se mostrará lo que la tienda ofrece al usuario para hacer la compra. Una vez elegidos los objetos a comprar, el programa mostrará el total de la compra por medio de la caja registradora, el vuelto que recibirá el comprador, y más cosas que aún se están pensando.

```python
def caja_registradora():
    num_digitos = int(input("¿Cuántos dígitos va a ingresar? "))
    plata = input("Ingrese la cantidad de dinero: ")
    
    while len(plata) != num_digitos:
        print(f"Error: debe ingresar {num_digitos} dígitos")
        plata = input("Ingrese la cantidad de dinero: ")
    
    print(" _______________________")
    print("|     Caja Registradora  |")
    print("|------------------------|")
    print("|        ________        |")
    print("|       |        |       |")
    print("|       |        |       |")
    print("|       |________|       |")
    print("|       __________       |")
    print("|      |          |      |")
    print("|      | $$$$$$$$ |      |")
    print("|      | $$$$$$$$ |      |")
    print("|      |__________|      |")
    print("|                        |")
    print("|________________________|")
    print("El dinero de la caja es $",plata)
caja_registradora() # llamada a la función para que se ejecute
```
